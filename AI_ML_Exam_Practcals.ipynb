```python
#(1).Explain BFS and DFS with a Python program.
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def bfs(start):
    queue = [start]
    visited = []

    while queue:
        node = queue.pop(0)
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            queue.extend(graph[node])

def dfs(start, visited=None):
    if visited is None:
        visited = []

    print(start, end=" ")
    visited.append(start)

    for n in graph[start]:
        if n not in visited:
            dfs(n, visited)

print("BFS:")
bfs('A')
print("\nDFS:")
dfs('A')
```

```python
#5).5. Write a calculator program using Python OOP (basic).
def calc(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        return "Invalid Operator"

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

print("Result:", calc(a, b, op))

```

```python
#6).Explain AI algorithms.

# 1. Linear Regression
from sklearn.linear_model import LinearRegression
print("\n--- Linear Regression ---")

X1 = [[1], [2], [3]]
y1 = [30000, 35000, 40000]

lr = LinearRegression()
lr.fit(X1, y1)

print("Prediction for 4 years experience:", lr.predict([[4]]))


# 2. Logistic Regression
from sklearn.linear_model import LogisticRegression
print("\n--- Logistic Regression ---")

X2 = [[30], [40], [50], [60]]
y2 = [0, 0, 1, 1]

log = LogisticRegression()
log.fit(X2, y2)

print("Prediction for age 45:", log.predict([[45]]))


# 3. Decision Tree
from sklearn.tree import DecisionTreeClassifier
print("\n--- Decision Tree ---")

X3 = [[5], [10], [15]]
y3 = ["Low", "Medium", "High"]

tree = DecisionTreeClassifier()
tree.fit(X3, y3)

print("Prediction for value 12:", tree.predict([[12]]))


# 4. K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier
print("\n--- KNN Algorithm ---")

X4 = [[1], [2], [3], [8], [9]]
y4 = ["Low", "Low", "Low", "High", "High"]

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X4, y4)

print("Prediction for value 4:", knn.predict([[4]]))


# 5. K-Means Clustering
from sklearn.cluster import KMeans
print("\n--- K-Means Clustering ---")

X5 = [[1], [2], [3], [10], [11], [12]]

kmeans = KMeans(n_clusters=2, n_init=10)
kmeans.fit(X5)

print("Cluster labels:", kmeans.labels_)


# 6. A* (A-Star) Pathfinding Algorithm
print("\n--- A* Algorithm ---")

def a_star(start, goal):
    open_list = [start]
    came_from = {}
    g = {start: 0}
    h = lambda x: abs(goal - x)   # simple heuristic
    f = {start: h(start)}

    while open_list:
        current = min(open_list, key=lambda x: f[x])

        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path.reverse()
            return path

        open_list.remove(current)

        # neighbors (simple: move +1 or -1)
        neighbors = [current - 1, current + 1]

        for n in neighbors:
            if n < 0:
                continue

            temp_g = g[current] + 1

            if n not in g or temp_g < g[n]:
                g[n] = temp_g
                f[n] = g[n] + h(n)
                came_from[n] = current
                if n not in open_list:
                    open_list.append(n)

    return None

path = a_star(2, 10)
print("Path from 2 to 10:", path)


print("\n--- END OF PROGRAM ---")

```
