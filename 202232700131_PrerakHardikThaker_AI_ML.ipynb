```python
#Q-1(a)
def calc(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        return "Invalid Operator"

a = float(input("Enter first digit: "))
b = float(input("Enter second digit: "))
op = input("Enter operator (+, -, *, /): ")

print("Result:", calc(a, b, op))


```

```python
#Q-2(a)
today_weather = {"city": "Ahmedabad", "temperature": 10, "sky": "cloudy", "precipitation": "rain"}
print("--- Today's Weather Fact ---")
print(today_weather)


rule_umbrella = {
    "conditions": [{"precipitation": "rain"}],
    "conclusion": {"action": "take umbrella"}
}

print("\n--- Applying the Rule To The Fact ---")
def apply_rule(fact, rule):
    for condition in rule["conditions"]:
        for key, value in condition.items():
            if key not in fact or fact[key] != value:
                return fact


    new_fact = fact.copy()
    new_fact.update(rule["conclusion"])
    return new_fact


inferred_weather = apply_rule(today_weather, rule_umbrella)

print("\n--- What we learned ---")
print(inferred_weather)
```

```python
#Q-3(a)
import collections

def bfs(graph, root):

    visited, queue = set(), collections.deque([root])
    visited.add(root)

    while queue:

        vertex = queue.popleft()
        print(str(vertex) + " ", end="")

        for neighbour in graph[vertex]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)


if __name__ == '__main__':
    graph = {0: [1, 2], 1: [2], 2: [3], 3: [1, 2]}
    print("Following is Breadth First Traversal: ")
    bfs(graph, 0)
```

```python
#Q-9(a)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

def bfs(start):
    queue = [start]
    visited = []

    while queue:
        node = queue.pop(0)
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            queue.extend(graph[node])

def dfs(start, visited=None):
    if visited is None:
        visited = []

    print(start, end=" ")
    visited.append(start)

    for n in graph[start]:
        if n not in visited:
            dfs(n, visited)

print("BFS:")
bfs('A')
print("\nDFS:")
dfs('A')
```
